```python
#############################
## Creates mock data       ##
## Fits a polynomial model ##
#############################

n = 100 # number of data points
x1 = np.linspace(0, 4, n) # yields 100 data point, evenly spaced between 0 and 4

y1 = x1 ** 3 - 6 * x1 ** 2 + 12 * x1 - 9
plt.scatter(x1, y1, marker = '.', color = 'black')
```

```python
######################################
## Makes a histogram of random data ##
######################################

noise = np.random.normal(0, 2, size = n) # get n noise values that have amplitudes roughly the same size as the data values themselves

plt.hist(noise, color = 'blue') # Gaussian with mean = 0, std = 2, 100 data points
```

```python
####################################################
## Plots the random data over the source function ##
####################################################

y2 = y1 + noise
plt.scatter(x1, y2, marker = '.', color = 'black')
plt.plot(x1, y1, color = 'blue')
```

```python
#####################################
## Creates mock measurement errors ##
#####################################

sigma = 1.5
y1_err = sigma * np.ones(n)
```

```python
###########################################################
## Fits the data twice- least squares and Simplex        ##
## Plots the random data, source function, and both fits ##
###########################################################

plt.errorbar(x1, y2, yerr = y1_err, color = 'black', fmt = '.', label = 'Random data')
plt.plot(x1, y1, color = 'blue', linewidth = 3, linestyle = '--', label = 'Actual function')

polymod = models.Polynomial1D(degree = 3) # initialize model as 3rd order polynomial
polyfit = fitting.LinearLSQFitter() # initialize the fitter as a linear least squares fitter
polybestfit = polyfit(polymod, x1, y2, weights = 1 / y1_err ** 2) # run fitter on data and errors
print(polybestfit)
print(polybestfit.parameters)

fitter_poly2 = fitting.SimplexLSQFitter()
polybestfit2 = fitter_poly2(polymod, x1, y2, weights = 1 / y1_err ** 2)

plt.plot(x1, polybestfit(x1), color = 'red', linewidth = 3, label = 'Best fit (Linear LSQ)')
plt.plot(x1, polybestfit2(x1), color = 'green', linewidth = 3, label = 'Bad fit (Simplex)')
plt.legend(loc = 'best')
```

The functional form of a sine function is:
$$f(x) = a\ \mathrm{sin}(bx+c)+d$$

```python
################################################
## 1. Create mock data                        ##
## Start by writing a function for a sinusoid ##
## Add noise and errors to the data           ##
## Create a function to create a plot         ##
################################################

def customsine(x, a, b, c, d):
    '''
    Takes an array of x values and the parameters for some sine function
    Returns y values
    '''
    yvals = a * np.sin(b * x + c) + d
    return yvals

n = 100
x = np.linspace(0, 3, n)
a = 5 # amplitude
b = 2 * np.pi # period
c = 0 # offest
d = 0 # offset

y = customsine(x, a, b, c, d)

noise = np.random.normal(0, 1, size = n) # mean = 0, std = 1
y_noise = y + noise

sigma = 1.5
y_err = sigma * np.ones(n)

def customplot(x, y, yerr = []):
    '''
    Convenience function to plot either a scatter or errorbar plot
    '''
    plt.figure()
    if len(yerr) == 0:
        plt.errorbar(x, y_noise, yerr = y_err, fmt = 'k.')
    else:
        plt.scatter(x, y_noise, color = 'black', marker = '.')
    plt.plot(x, y, color = 'blue')
    plt.draw()
    return None

customplot(x, y)
```

```python
##################################################
## 2. Fit it                                    ##
## Defines the desired fitter                   ##
##                                              ##
## 3. Make the plots                            ##
## Plots the random data with the best fit line ##
##################################################

from astropy.modeling import fitting, models, functional_models

sinemod = functional_models.Sine1D()
fitter_sine = fitting.LevMarLSQFitter() # better fitter for sine data
bestfit_sine = fitter_sine(sinemod, x, y, weights = 1 / y_err ** 2)
print(bestfit_sine.parameters) # given parameters

fit_yvals = bestfit_sine(x)
print(fit_yvals)

plt.figure()
plt.errorbar(x, y_noise, yerr = y_err, fmt = 'k.')
plt.plot(x, fit_yvals, color = 'blue', linewidth = 3)
```

```python
################################################################################
## 4. Store the data in pandas dataframes                                     ##
## Put everything into a dataframe and save it so we can pass it on to others ##
################################################################################

array_for_dataframe = [x, y_noise, y_err, fit_yvals]
array_for_dataframe = np.array(array_for_dataframe).T # transpose from 100 columns and 4 rows to 4 columns and 100 rows

print(array_for_dataframe)
```

```python
########################################
## Initializes a new pandas dataframe ##
## Stores our data in the dataframe   ##
########################################

new_df = pd.DataFrame(array_for_dataframe)
new_df
```

```python
###############################################
## Assigns new column names to the dataframe ##
###############################################

column_names = ['Time', 'Amplitude', 'Amplitude Error', 'Best Fit Value']
new_df.columns = column_names
new_df
```

```python
##################################################
## 5. Save and deliver the plots and dataframes ##
## Saves all data to the pandas dataframe       ##
##################################################

new_df.to_csv('sine_data_from_experiment.csv')
```
